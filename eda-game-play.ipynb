# Step 3 - EDA

```python
from wordcloud import WordCloud
```

```python
#Visualisation
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Start with the eventname column
text = " ".join(x for x in train["event_name"])

# Create and generate a word cloud image:
wordcloud = WordCloud(width = 800, height = 250, 
            background_color ="black",colormap="RdYlGn",max_font_size=100, stopwords =None,repeat= True).generate(text)
plt.figure(figsize = (10, 8),facecolor= "#254441") 

# Display the generated image:
plt.imshow(wordcloud)
plt.axis("off")
plt.margins(x=0, y=0)
plt.tight_layout(pad = 0) 
plt.show()
```

4 most frequently event_type appear in the train dataset are:
* cutscene_click
* navigate_click
* object_click
* person_click

Other than click event type, people tend to hover over the map and click into it as well  (event_type = map_hover, map_click)

* cutscene_click 
* navigate_click 43%
* object_click
* person_click 23%

48% of all events are basic , 48% are undefined, and 4% are others.

```python
# distribution for level column
plt.hist(train['level'], bins=20)
plt.show()
```

```python
# distribution for page column
plt.hist(train['page'], bins=5)
plt.show()
```

```python
# distribution for level group column
plt.hist(train['level_group'], bins=5)
plt.show()
```

A histogram is a value distribution plot of numerical columns. It basically creates bins in various ranges in values and plots it where we can visualize how values are distributed. We can have a look where more values lie like in positive, negative, or at the center(mean).
